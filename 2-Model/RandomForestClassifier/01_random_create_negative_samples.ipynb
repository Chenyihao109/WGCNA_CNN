```python
interaction_df_negative=pd.DataFrame(columns=["node1","node2","label"])
for n in range(6000):
    negative_sample_node1=protein_map_list[np.random.randint(798)]
    negative_sample_node2=protein_map_list[np.random.randint(798)]
    flag1=np.where(node1==negative_sample_node1)  #判断已知的关系中，节点1集合中是否出现生成的负样本的第一个结点
    flag2=np.where(node2==negative_sample_node1)#判断已知的关系中，节点2集合中是否出现生成的负样本的第一个结点
    flag1_int=flag1[0].size
    flag2_int=flag2[0].size
    if flag1_int==0 and flag2_int==0:
        interaction_df_negative=interaction_df_negative._append({"node1":negative_sample_node1,"node2":negative_sample_node2,"label":0},ignore_index=True)
        continue
    elif flag1_int:
        r=range(0,flag1_int)
        for i in r:
            if node2[flag1[0][i]]==negative_sample_node2:
                break
            elif i==flag1_int-1:
               interaction_df_negative=interaction_df_negative._append({"node1":negative_sample_node1,"node2":negative_sample_node2,"label":0},ignore_index=True)
    else:
        r=range(0,flag2_int)
        for i in r:
            if node1[flag2[0][i]]==negative_sample_node1:
                break
            elif i==flag2_int-1:
               interaction_df_negative=interaction_df_negative._append({"node1":negative_sample_node1,"node2":negative_sample_node2,"label":0},ignore_index=True)

    
```
